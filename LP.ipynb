```python

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

arr = ['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
for col in arr:
    df1[col]=label_encoder.fit_transform(df1[col])
```

```python
x = df1.drop(['Loan_ID','Loan_Status'],axis=1)
y = df1['Loan_Status'] 
```

```python
from sklearn.model_selection import train_test_split    
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=0)    
```

```python
score = []
```

```python
score
```

```python
from sklearn.tree import DecisionTreeRegressor

model1 = DecisionTreeRegressor(random_state=1)
model1.fit(x_train,y_train)
score.append(model1.score(x_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression  
classifier= LogisticRegression(random_state=0)  
classifier.fit(x_train, y_train) 
score.append(classifier.score(x_test,y_test))
```

```python
from sklearn.svm import SVC # "Support vector classifier"  
classifier2 = SVC(kernel='linear', random_state=0)  
classifier2.fit(x_train, y_train)  
score.append(classifier2.score(x_test,y_test))
```

```python
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(x_train, y_train) 
score.append(regressor.score(x_test,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier(random_state=1)
model2.fit(x_train,y_train)
score.append(model2.score(x_test,y_test))
```

```python
import seaborn as sns
sns.catplot(x="Gender",y="Married",
           hue="Loan_Status",
           kind="bar",
           data=df1)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
score.append(RFC.score(x_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(x_train,y_train)
score.append(neigh.score(x_test,y_test))
```

```python
d = {'Name':['DecisionTreeRegressor','LogisticRegression','SVC','Linear','DTC','RFC','KNC'],
    'Accuracy':score}
new = pd.DataFrame(d)
```
